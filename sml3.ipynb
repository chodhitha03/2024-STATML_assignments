**Question 2:**

**Find log-likelihood values for 50, 75, and 80. For California housing [‘latitude’] and find the Max likelihood values.**

```python
mu = np.mean(a['latitude'])
sigma = np.var(a['latitude'], ddof=1)

def log_likelihood(x, mu, sigma):

    return -0.5 * (np.log(2 * np.pi * sigma**2) + ((x - mu)**2 / sigma**2))

values = [50, 75, 80]
log_likelihoods = [log_likelihood(v, mu, sigma) for v in values]

print("Mean :", mu)
print("Variance:", sigma)
print("Log-Likelihoods for values [50, 75, 80]:", log_likelihoods)
```

**Question 3**

Estimate the density of each column of california_housing [‘latitude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, For the bins [5, 10, 15, 20]. And write the observations.

```python
import matplotlib.pyplot as plt
columns = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']
bins = 5

for column in columns:
    plt.figure(figsize=(10, 6))
    plt.hist(a[column], bins=bins, density=True, alpha=0.6, color='blue', edgecolor='black')
    plt.title(f'Density Plot for {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
```
