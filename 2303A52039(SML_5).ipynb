*4.Convert string valued features to numbers*


```python
import pandas as pd
```

```python
a=pd.read_csv('/content/diabetes_data_upload.csv')
print(a)
```

```python
y=a['class']
print(y)
```

```python
x=a.drop('class',axis=1)
print(x)
x=x.fillna(0)

print(x)
```

```python
mapping={'Positive':1,'Negative':0}
y=y.map(mapping)
print(y)
```

```python
mapping={'Male':1,'Female':0}
x['Gender']=x['Gender'].map(mapping)
print(x)
```

```python
mapping={'No':0,'Yes':1}
x=x.replace('Yes',1)
x=x.replace('No',0)
print(x)
```

```python
print(x.isnull().sum())
```

```python
import matplotlib.pyplot as plt
```

```python
for column in x.columns:
  plt.plot(x[column],y)
  plt.xlabel(column)
  plt.ylabel('class')
  plt.show()
  plt.title(f'Scatter plot of {column} vs class')
```
