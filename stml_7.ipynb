From the above data:

Read the data with pandas and find features and target variables

FInd MSE, MAE, for test size (20,25)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('/content/Salary_Data[1].csv')
d
```

```python
X = d[['YearsExperience']]
y = d['Salary']
```

```python
plt.scatter(X, y, color='blue')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.plot(X, y)
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
for test_size in [0.20, 0.25]:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f"Test Size: {int(test_size*100)}%")
    print(f"MSE: {mse}")
    print(f"MAE: {mae}\n")

    plt.scatter(X_test, y_test, color='blue', label='Test Data')
    plt.plot(X_test, y_pred, color='red', label='Best Fit Line')
    plt.title(f'Linear Regression with Test Size {int(test_size*100)}%')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
print(d)
print("")
d.head()

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)

print(f"MSE: {mse}\n")
print(f"MAE: {mae}\n")
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Best fit line')
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = {
    'Feature1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

d = pd.DataFrame(data)
```

```python
X = d[['Feature1', 'Feature2', 'Feature3', 'Feature4']]
y = d['Target']

plt.scatter(d['Feature1'], d['Target'], color='blue', label='Feature 1 vs Target')
plt.title('Feature 1 vs Target')
plt.xlabel('Feature 1')
plt.ylabel('Target (Y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(X, y)
plt.title('Features vs Target')
plt.xlabel('Features')
plt.ylabel('Target')
plt.show()
```

```python
for test_size in [0.20, 0.30]:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f"Test Size: {int(test_size*100)}%")
    print(f"MSE: {mse}")
    print(f"MAE: {mae}\n")

    plt.scatter(X_test['Feature1'], y_test, color='blue', label='Test Data')
    plt.plot(X_test['Feature1'], y_pred, color='red', label='Best Fit Line')
    plt.title(f'Linear Regression with Test Size {int(test_size*100)}%')
    plt.xlabel('Feature 1')
    plt.ylabel('Target (Y)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
X = d[['Feature1', 'Feature2', 'Feature3', 'Feature4']]
y = d['Target']
print(d)
print("")

d.head()

model = LinearRegression()
model.fit(X, y)


y_pred = model.predict(X)


mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
print(f"MSE: {mse}")
print(f"MAE: {mae}\n")

plt.scatter(d['Feature1'], y, color='blue', label='Actual data')
plt.plot(d['Feature1'], y_pred, color='red', label='Best fit line')
plt.title('Feature1 vs Target')
plt.xlabel('Feature1')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```
